```python
#חישוב עמודת תקציב כולל מע״מ (18%)
df['budget_with_vat'] = df['budget'] * 1.18
df['budget_with_vat'].head()
```

```python
#חישוב עמודת הכנסות כולל מע״מ (18%)
df['revenue_with_vat'] = df['revenue'] * 1.18
df['revenue_with_vat'].head()
```

```python
#הצגת חמש הפופולריות הנמוכות ביותר
print(df['popularity'].nsmallest(5))
```

```python
#הצגת תאריכי הוספה של הסרטים
print(df['date_added'])
```

```python
#בדיקת סוג נתונים בעמודת תאריך ההוספה
type(df['date_added'])
```

```python
#נמיר את התאריך לDATE כי הוא ב OBJECT
df['date_added'] = pd.to_datetime(df['date_added'], format="%Y-%m-%d")
```

```python
#חילוץ שנה מתאריך
df['year_date_added'] = df['date_added'].dt.year
print("כמה סרטים נוספו בכל שנה***************88")
print(df['year_date_added'].value_counts().sort_index())
```

```python
#חילוץ חודש מתאריך
df['month_date_added']=df['date_added'].dt.month
df.info()
print(df.isnull().sum())
```

```python
#טיפול בעמודת duration – חסר ב־100% מהשורות (כל 16,000)
#נמחק את כל העמודה כי אם אין נתונים -לא שמיש
df.drop(columns=['duration'], inplace=True)
print(df.columns)
```

```python
#טיפול בעמודות director, cast, description, genres-נמלא את הערכים החסרים בUnknown
for i in ['director', 'cast', 'description', 'genres']:
    df[i].fillna('Unknown', inplace=True)
print(df.isnull().sum())
```

```python
#נטפל ב country country – חסר ב־466 שורות
df["country"].fillna("Unknown", inplace=True)
print(df.isnull().sum())
print("\n--- פופולריות ממוצעת לפי סוג ---")
print(df.groupby('type')['popularity'].mean())
```

```python
#מציג את ההכנסות הממוצעות לפי שנה ואת כמות הפעמים שכל ז'אנר הופיע לפי שנה.
pd.options.display.float_format = '${:,.0f}'.format
yearly_revenue = df[df['revenue'] > 0].groupby('release_year')['revenue'].mean().sort_index()
print("הכנסות ממוצעות לפי שנה:")
print(yearly_revenue)
genres_by_year = (
    df[['release_year', 'genres']]
    .assign(genres=df['genres'].str.split(', '))
    .explode('genres') .groupby(['release_year', 'genres']) .size() .reset_index(name='count')
    .sort_values(['release_year', 'count'], ascending=[True, False]).sort_index())
print(genres_by_year)
```

```python
#הסרט הכי רווחי והסרט עם הכי הרבה דירוגים לפי שנה
most_profitable_movie = df[df['revenue'] == df['revenue'].max()][['title', 'revenue']]
print("הסרט הכי רווחי בכל הזמנים:")
print(most_profitable_movie)
most_voted_by_year = df.loc[
    df.groupby('release_year')['vote_count'].idxmax()][['release_year', 'title', 'vote_count']].sort_index()
print("הסרט עם הכי הרבה דירוגים בכל שנה:")
print(most_voted_by_year)
```

```python
# סינון סרטים עם תקציב והכנסה תקינים
df_profit_check = df[(df['budget'] > 0) & (df['revenue'] > 0)]
df_profit_check
```

```python
# חישוב רווח: הכנסה פחות תקציב
profit_amount = df_profit_check['revenue'] - df_profit_check['budget']
profit_amount
```

```python
# חישוב סטטוס רווחיות
profit_status = ['רווחי' if profit >= 0 else 'לא רווחי' for profit in profit_amount]
```

```python
# יצירת טבלה חדשה עם שם הסרט, רווח וסטטוס
result = pd.DataFrame({
    'title': df_profit_check['title'],
    'רווח (ש"ח)': profit_amount,
    'סטטוס': profit_status
})
print(result.head(50))
```

```python
# סינון רק של סרטים לא רווחיים
not_profitable = result[result['סטטוס'] == 'לא רווחי']
print(not_profitable)
```

```python
#הצגת 10 הסרטים הרווחיים ביותר
profitable = result[result['סטטוס'] == 'רווחי'].sort_values('רווח (ש"ח)', ascending=False).sort_index()
print(profitable.head(10))
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
# ספירת סרטים לפי מדינה
movies_by_country = df[df['type'] == 'Movie']['country'].value_counts().head(10).reset_index()
movies_by_country.columns = ['country', 'count']
movies_by_country
```

```python
# גרף barplot עם צבעים
plt.figure(figsize=(10, 6))
sns.barplot(data=movies_by_country, x='count', y='country', hue='country')
plt.title("Top 10 Countries by Number of Movies")
plt.xlabel("Number of Movies")
plt.ylabel("Country")
plt.show()
```

```python
#היסטוגרמת הפצה של ממוצע הדירוגים
sns.histplot(df['vote_average'], bins=20, kde=True)
plt.title("Distribution of Vote Average")
plt.xlabel("Vote Average")
plt.ylabel("Count")
plt.show()
```

```python
#הצגת גרף עמודות 10 הסרטים עם הכי הרבה דירוגים
top_10_movies = df.sort_values('vote_count', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(data=top_10_movies, x='vote_count', y='title')
plt.title("Top 10 Highest Rated Movies")
plt.xlabel("Vote ")
plt.ylabel("Movie")
plt.show()
```

```python
# # גרף: דירוג מול הכנסה
import matplotlib.ticker as ticker
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='vote_average', y='revenue')
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
plt.title("Relationship between rating and average")
plt.xlabel("Average rating")
plt.ylabel("income")
plt.tight_layout()
plt.show()
```

```python
#סינון סרטים לפי ז'אנר ושנה
df['main_genre'] = df['genres'].str.split(',').str[0]
genre_order = df[df['type'] == 'Movie']['main_genre'].value_counts().index
top_years = df['release_year'].value_counts().head(10).index
filtered = df[(df['release_year'].isin(top_years)) & (df['type'] == 'Movie')]
filtered
```

```python
#הצגת מספר הסרטים לפי ז'אנר ושנה לפי גרף עמודות
plt.figure(figsize=(12, 10))
sns.countplot(data=filtered, y='main_genre', hue='release_year')
plt.title("Number of films by main genre and by year")
plt.xlabel("Quantity of films")
plt.ylabel("genre")
plt.legend(title='שנה', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# גרף פאי לפי שפת הסרטים (language)
lang_counts = df['language'].value_counts().head(6)  # 6 השפות הכי נפוצות
plt.figure(figsize=(8, 8))
plt.pie(lang_counts, labels=lang_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set3"))
plt.title("Distribution of Movie Languages (Top 6)")
plt.axis('equal')
plt.show()
```
